```python
y = 0
lst = []
x = float(input('Input : '))
lst.append(x)
while x != 0.0:
    x = float(input('Input : '))
    lst.append(x)
    y += 1
print(y)
avg = sum(lst)/y
a = avg*y
print('Mean : {}, Sum : {}'.format(avg,a))
```

```python
x = int(input('Input upper bound : '))
y = int(input('Input lower bound : '))
z = list(range(x,y+1,))
w = 0

for i in z:
    for j in range(2,i+1):
            if i%j == 0:
                w += 1
            else: w += 0
    if w == 1:
        print('Prime Number is ', i)
    w = 0
```
