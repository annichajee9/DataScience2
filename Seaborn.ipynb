```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
Name = ['A', 'B', 'C', 'D']
Year = ['2014', '2015', '2016', '2017']
Income = np.random.uniform(50, 500, (4,4))
df = pd.DataFrame(Income, Name, Year)
df
```

```python
sns.barplot(x = Name, y = df.mean(axis = 1))
```

```python
sns.barplot(y = Name, x = df.mean(axis = 1))
```

```python
sns.scatterplot(x=df.index, y = df['2017'], data = df)
```

```python
new_df1 = pd.crosstab(df1['Pclass'], df1['Sex'])
print(new_df1)
sns.heatmap(new_df1)
```

```python
new2_df1 = df1.pivot_table(index='Pclass', columns='Sex', values='Fare')
print(new2_df1)
sns.heatmap(new2_df1, annot=True, cmap='coolwarm')
```
